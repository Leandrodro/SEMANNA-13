```python
import numpy as np

# Ejercicio 1: Valor presente de una anualidad con gradiente uniforme
def ejercicio1():
    n = 8  # Número de periodos
    i = 0.05  # Tasa efectiva mensual
    C = 100  # Cuota base
    Cuota_final = 240  # Cuota final

    incremento_total = Cuota_final - C
    gradiente = incremento_total / (n - 1)
    
    VP = gradiente * ((1 - (1 + i) ** -n) / i**2 - (n / (i * (1 + i)**n)))
    print(f"Ejercicio 1: El valor presente de la anualidad es: {VP:.2f}")

# Ejercicio 2: Monto total acumulado
def ejercicio2():
    C = 500  # Depósito inicial
    G = 100  # Incremento mensual
    n = 12  # Número de periodos
    i = 0.03  # Tasa de interés mensual
    
    # Cálculo del aporte fijo acumulado
    MC = C * ((1 + i)**n - 1) / i
    
    # Cálculo del gradiente acumulado
    MG = G * (((1 + i)**n - 1) / i**2 - n / (i * (1 + i)**n))
    
    # Monto total acumulado
    M = MC + MG
    print(f"Ejercicio 2: El monto total acumulado es: {M:.2f}")

# Ejercicio 3: Valor presente total incluyendo mantenimiento
def ejercicio3():
    C = 100  # Cuota inicial
    G = 20  # Incremento mensual
    n = 12  # Número de periodos
    i = 0.04  # Tasa efectiva mensual
    VP_inicial = 4000  # Costo inicial

    VPC = C * (1 - (1 + i)**-n) / i
    VPG = G * (((1 - (1 + i)**-n) / i**2) - n / (i * (1 + i)**n))
    
    VP = VP_inicial + VPC + VPG
    print(f"Ejercicio 3: El valor presente total es: {VP:.2f}")

# Ejercicio 4: Cálculo del importe del gradiente
def ejercicio4():
    S = 10000  # Valor futuro
    R = 500  # Cuota base
    n = 10  # Número de periodos
    i = 0.03  # Tasa efectiva mensual

    # Fórmula para calcular el gradiente
    parte_gradiente = (1 / i) * (((1 + i)**n - 1) / i - n)
    parte_renta = R * ((1 + i)**n - 1) / i
    G = (S - parte_renta) / parte_gradiente
    
    print(f"Ejercicio 4: El importe del gradiente es: {G:.2f}")

# Ejercicio 5: Importe de la última cuota
def ejercicio5():
    P = 10000  # Valor presente del préstamo
    G = 100  # Incremento mensual
    n = 12  # Número de periodos
    i = 0.05  # Tasa efectiva mensual

    Pg = (G / i) * (((1 + i)**n - 1) / ((i) * (1 + i)**n) - n / (1 + i)**n)
    Pr = P - Pg
    R = Pr * (i * (1 + i)**n) / ((1 + i)**n - 1)
    ultima_cuota = R + (n - 1) * G
    
    print(f"Ejercicio 5: El importe de la última cuota es: {ultima_cuota:.2f}")

# Ejercicio 6: Costo mensual equivalente
def ejercicio6():
    P = 2000  # Desembolso inicial
    R = 50  # Cuota base
    G = 20  # Incremento mensual
    n = 10  # Número de periodos
    i = 0.05  # Tasa efectiva mensual

    Pr = R * ((1 + i)**n - 1) / (i * (1 + i)**n)
    Pg = (G / i) * (((1 + i)**n - 1) / (i * (1 + i)**n) - n / (1 + i)**n)
    R_total = (P + Pr + Pg) * (i * (1 + i)**n) / ((1 + i)**n - 1)
    
    print(f"Ejercicio 6: El costo mensual equivalente es: {R_total:.2f}")

# Ejercicio 7: Importe del gradiente en anualidad creciente
def ejercicio7():
    n = 10  # Número de periodos
    R = 500  # Cuota base
    VP = 5000  # Valor presente
    TET = 0.0927  # Tasa trimestral efectiva

    VPC = R * ((1 - (1 + TET)**-n) / TET)
    VPG = VP - VPC
    G = VPG * TET**2 / (((1 + TET)**n - 1) - n * TET * (1 + TET)**-n)
    
    print(f"Ejercicio 7: El importe del gradiente es: {G:.2f}")

# Ejercicio 8: Renta mensual uniforme equivalente
def ejercicio8():
    flujos = [60, 60, 80, 100, 120, 140, 160, 180, 200]
    i = 0.05  # Tasa efectiva mensual
    
    VP = sum(f / (1 + i)**(t + 1) for t, f in enumerate(flujos))
    R = VP * (i * (1 + i)**len(flujos)) / ((1 + i)**len(flujos) - 1)
    
    print(f"Ejercicio 8: La renta mensual uniforme equivalente es: {R:.2f}")

# Ejercicio 9: Valor presente de anualidad con gradiente negativo
def ejercicio9():
    C = 1000  # Cuota base
    G = -50  # Gradiente
    n = 20  # Número de periodos
    TET = 0.1576  # Tasa trimestral efectiva

    VPC = C * ((1 - (1 + TET)**-n) / TET)
    VPG = G * (((1 - (1 + TET)**-n) / TET**2) - n / (TET * (1 + TET)**n))
    VP = VPC + VPG
    
    print(f"Ejercicio 9: El valor presente de la anualidad es: {VP:.2f}")

# Ejercicio 10: Valor presente de préstamo con cuotas crecientes
def ejercicio10():
    C = 500  # Cuota base
    g = 0.05  # Incremento relativo
    n = 24  # Número de periodos
    TET = 0.0927  # Tasa trimestral efectiva

    VP = C / (1 + TET) * ((1 + g)**n - (1 + TET)**n) / (g - TET)
    print(f"Ejercicio 10: El valor presente del préstamo es: {VP:.2f}")

# Ejecución de todos los ejercicios
ejercicio1()
ejercicio2()
ejercicio3()
ejercicio4()
ejercicio5()
ejercicio6()
ejercicio7()
ejercicio8()
ejercicio9()
ejercicio10()
```
